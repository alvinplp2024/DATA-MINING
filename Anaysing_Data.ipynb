# Apply Function()

```python
#Classify functions based on ratings
def rating_group (rating):
    if rating >= 7.5:
         return "Good"
    elif rating >= 6.0:
        return "Average"
    else:
        return "Bad"
```

```python
#Creating  a new variable ' rating catwegory ' to hold rating classification 
data['Rating_category']= data['Rating'].apply(rating_group)
```

```python
data[['Title','Director','Rating','Rating_category']].head(5)
```
